```python
df = pd.read_csv('data_for_model.csv')
df.head(5)
```

```python
df=df.drop(columns=['bedrooms','bathrooms','balcony'])
```

```python
# initialize Encoder
encoder = LabelEncoder()

# Fit and transform the categorical columns
df['flat_type'] = encoder.fit_transform(df['flat_type'])
df['location1'] = encoder.fit_transform(df['location1'])
df['age_of_property'] = encoder.fit_transform(df['age_of_property'])
df['furnishing'] = encoder.fit_transform(df['furnishing'])
df['parking'] = encoder.fit_transform(df['parking'])
df.head(5)
```

```python
# split into dependent and indipendent variable
X = df.drop('flat_price_in_crore',axis=1)
y = df['flat_price_in_crore']
```

```python
# split data into train and test
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Fit on MLR Model
LinearRegression()

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# check and verify all error and prdict and score
predict = model.predict(X_test)
r2 = r2_score(y_test,predict)
print(r2)
```
